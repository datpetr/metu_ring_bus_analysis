```python
def combine_departments(row):
    """Combine department columns, filtering out NaN values"""
    departments = [str(val) for val in row if pd.notna(val)]
    return '; '.join(departments) if departments else None
```

```python
# ============================================================
# 8. Verify all Turkish characters are converted (except open-ended)
# ============================================================

# Define open-ended columns that may legitimately contain Turkish text
open_ended_columns = ['Affiliation_Program', 'Open_Comments']

# Turkish character set
turkish_chars = set('çğıöşüÇĞİÖŞÜ')

# Check each column for Turkish characters
print("Checking for remaining Turkish characters:\n")
print("=" * 70)

has_turkish = False
for col in df_merged.columns:
    if col in open_ended_columns:
        print(f"⊘ {col:<40} (SKIP - open-ended column)")
        continue
    
    # Check if column contains Turkish characters
    turkish_found = []
    for idx, val in df_merged[col].items():
        if pd.notna(val):
            str_val = str(val)
            if any(char in str_val for char in turkish_chars):
                turkish_found.append((idx, str_val[:50]))
    
    if turkish_found:
        has_turkish = True
        print(f"✗ {col:<40} CONTAINS TURKISH ({len(turkish_found)} rows)")
        for idx, sample in turkish_found[:3]:  # Show first 3 examples
            print(f"    Row {idx}: {sample}")
    else:
        print(f"✓ {col:<40} OK")

print("=" * 70)
if not has_turkish:
    print("\n✓ SUCCESS: All data (except open-ended) is in English!")
else:
    print("\n✗ WARNING: Some columns still contain Turkish characters.")
    print("   Review the mapping above and add missing translations.")
```

```python
# ============================================================
# 7b. Handle multi-value columns (comma-separated)
# ============================================================

def translate_multi_value(value, mapping_dict):
    """Translate comma-separated values"""
    if pd.isna(value):
        return value
    
    str_val = str(value)
    # Split by comma, translate each part, rejoin
    parts = [part.strip() for part in str_val.split(',')]
    translated_parts = [mapping_dict.get(part, part) for part in parts]
    return ', '.join(translated_parts)

# Define which columns have multi-value responses
multi_value_columns = [
    'Weekday_Ring_TimeSlots',
    'Weekend_Ring_Frequency',
    'Weekend_Ring_TimeSlots',
    'Route_Used_Most',
    'Ring_Purposes',
    'Ring_Reasons',
    'Residence'
]

# Translate multi-value columns
for col in multi_value_columns:
    if col in df_merged.columns:
        df_merged[col] = df_merged[col].apply(
            lambda x: translate_multi_value(x, turkish_to_english)
        )

print("✓ Multi-value columns translated!")

# Now run the verification again
print("\nChecking for remaining Turkish characters:\n")
print("=" * 70)

open_ended_columns = ['Affiliation_Program', 'Open_Comments']
turkish_chars = set('çğıöşüÇĞİÖŞÜ')

has_turkish = False
for col in df_merged.columns:
    if col in open_ended_columns:
        print(f"⊘ {col:<40} (SKIP - open-ended column)")
        continue
    
    turkish_found = []
    for idx, val in df_merged[col].items():
        if pd.notna(val):
            str_val = str(val)
            if any(char in str_val for char in turkish_chars):
                turkish_found.append((idx, str_val[:50]))
    
    if turkish_found:
        has_turkish = True
        print(f"✗ {col:<40} CONTAINS TURKISH ({len(turkish_found)} rows)")
        for idx, sample in turkish_found[:2]:
            print(f"    Row {idx}: {sample}")
    else:
        print(f"✓ {col:<40} OK")

print("=" * 70)
if not has_turkish:
    print("\n✓ SUCCESS: All data (except open-ended) is in English!")
else:
    print("\n✗ Still some Turkish found - check samples above")

print(f"Shape: {df_merged.shape}")
```
