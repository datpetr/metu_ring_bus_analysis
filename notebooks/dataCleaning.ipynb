```python
usage_cols = [c for c in df.columns if "usage" in c.lower() or "frequency" in c.lower()]
satisfaction_cols = [c for c in df.columns if "satisfaction" in c.lower()]
```

```python
# ============================================================
# CONVERT SATISFACTION COLUMNS TO NUMERIC (UPDATED MAPPING)
# ============================================================

# Define the satisfaction mapping - INCLUDING ALL VARIATIONS FOUND IN YOUR DATA
satisfaction_map = {
    # Standard Likert scale
    "Very dissatisfied": 1,
    "Dissatisfied": 2,
    "Neutral": 3,
    "Satisfied": 4,
    "Very satisfied": 5,
    
    # Variations found in your data
    "Very Dissatisfied": 1,
    "Very Satisfied": 5,
    "Unsatisfied": 2,              # ADD THIS - maps to Dissatisfied (2)
    "Very Unsatisfied": 1,        # ADD THIS - maps to Very dissatisfied (1)
    "Highly Satisfied": 5,        # ADD THIS - maps to Very satisfied (5)
}

# Check what values actually exist BEFORE mapping
print("Original values in satisfaction columns (BEFORE mapping):")
for col in satisfaction_cols:
    if col in df.columns and not pd.api.types.is_numeric_dtype(df[col]):
        print(f"\n{col}:")
        print(df[col].value_counts(dropna=False))

# Reload the dataframe to start fresh (or work with original values)
# If you've already mapped, you might need to reload:
# df = pd.read_csv('../data/merged_metu_ring.csv')

# Apply mapping
for col in satisfaction_cols:
    if col in df.columns:
        # Check if already numeric
        if pd.api.types.is_numeric_dtype(df[col]):
            print(f"✓ {col} is already numeric - skipping")
        else:
            # Store original non-null count
            original_non_null = df[col].notna().sum()
            
            # Convert to numeric
            df[col] = df[col].map(satisfaction_map)
            
            # Check for unmapped values
            new_null_count = df[col].isna().sum()
            original_null_count = df[col].isna().sum() - (original_non_null - df[col].notna().sum())
            
            if df[col].isna().any():
                # Find what values couldn't be mapped
                # We need to check original values - but if already converted, reload first
                print(f"⚠ {col}: {new_null_count} NaN values after mapping")
            else:
                print(f"✓ Converted {col} to numeric successfully")

# Verify conversion
print("\n" + "="*60)
print("Verification - Data types and NaN counts after conversion:")
for col in satisfaction_cols:
    if col in df.columns:
        nan_count = df[col].isna().sum()
        print(f"{col:40} dtype: {df[col].dtype}, NaN: {nan_count}, range: [{df[col].min():.0f}, {df[col].max():.0f}]")
```

```python
# Only calculate variance on numeric columns
likert_cols = [c for c in satisfaction_cols if pd.api.types.is_numeric_dtype(df[c])]

if len(likert_cols) == 0:
    print("⚠ No numeric satisfaction columns found! Run the conversion cell above first.")
else:
    print(f"Calculating variance for: {likert_cols}")
    df["likert_variance"] = df[likert_cols].var(axis=1)
    
    straight_liners = df[df["likert_variance"] == 0]
    print(f"Found {len(straight_liners)} straight-liners out of {len(df)} total")
```
